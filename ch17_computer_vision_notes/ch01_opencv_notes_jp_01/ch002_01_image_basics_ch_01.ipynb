# resize the images

```python
plt.imshow(fix_img)
```

```python
fix_img.shape
```

```python
re_fix_img=cv2.resize(fix_img,(1000,800))
plt.imshow(re_fix_img)
```

```python
re_fix_img=cv2.resize(fix_img,(100,100))
plt.imshow(re_fix_img)
```

```python
w_ratio=0.5
h_ratio=0.5

new_img=cv2.resize(fix_img,(0,0),fix_img,w_ratio,h_ratio)
plt.imshow(new_img)
print(new_img.shape)
```

# flip images

```python
#flip images 

new_img1=cv2.flip(fix_img,0)
plt.imshow(new_img1)
```

```python
#flip images 

new_img2=cv2.flip(fix_img,1)
plt.imshow(new_img2)
```

```python
#flip images 

new_img3=cv2.flip(fix_img,-1)
plt.imshow(new_img3)
```

```python
#write an image 

cv2.imwrite("new_img_puppy.jpg",new_img3)
```

# enlarge

```python
fig=plt.figure(figsize=(10,8))
ax=fig.add_subplot(111)
ax.imshow(fix_img)
```
