# blending 

# blending on both images are same size 

```python
import cv2 

img1=cv2.imread("DATA/dog_backpack.png")
img1=cv2.cvtColor(img1,cv2.COLOR_BGR2RGB)

img2=cv2.imread("DATA/watermark_no_copy.png")
img2=cv2.cvtColor(img2,cv2.COLOR_BGR2RGB)
```

```python
img1.shape
```

```python
img2.shape
```

```python
#both are different and convert into same size 

img1=cv2.resize(img1,(600,600))
img2=cv2.resize(img2,(600,600))

plt.imshow(img1)
```

```python
plt.imshow(img2)
```

```python
# addWeighted method will work only same size of images 

blended=cv2.addWeighted(src1=img1, alpha=0.5, src2=img2, beta=0.5, gamma=0)
plt.imshow(blended)
```

```python
blended=cv2.addWeighted(src1=img1, alpha=0.9, src2=img2, beta=0.17, gamma=0)
plt.imshow(blended)
```

# different images 

```python
import cv2 

img1=cv2.imread("DATA/dog_backpack.png")
img1=cv2.cvtColor(img1,cv2.COLOR_BGR2RGB)

img2=cv2.imread("DATA/watermark_no_copy.png")
img2=cv2.cvtColor(img2,cv2.COLOR_BGR2RGB)
```

```python
img2=cv2.resize(img2,(600,600))
```

```python
plt.imshow(img1)
```

```python
img1.shape
```

```python
x_offset=img1.shape[1]-img2.shape[1]
#x_offset
y_offset=img1.shape[0]-img2.shape[0]
#y_offset
```

```python
img1.shape
```

```python
x_offset
```

```python
y_offset
```

```python
roi=img1[y_offset:img1.shape[0],x_offset:img1.shape[1]]
plt.imshow(roi)
```

```python
img2togray=cv2.cvtColor(img2,cv2.COLOR_RGB2GRAY)
plt.imshow(img2togray,cmap="gray")
```

```python
masked_inv=cv2.bitwise_not(img2togray)
plt.imshow(masked_inv,cmap="gray")
```

```python
masked_inv.shape
```

```python
import numpy as np 

white_background=np.full(img2.shape,255,dtype=np.uint8)
white_background.shape
```

```python
bk=cv2.bitwise_or(white_background,white_background,mask=masked_inv)
bk.shape
```

```python
plt.imshow(bk)
```

```python
fg=cv2.bitwise_or(img2,img2,mask=masked_inv)
plt.imshow(fg)
```

```python
final_roi=cv2.bitwise_or(roi,fg)
plt.imshow(final_roi)
```

```python
large_image=img1
small_image=final_roi
```

```python
large_image[y_offset:y_offset+small_image.shape[0],x_offset:x_offset+small_image.shape[1]]=small_image
```

```python
plt.imshow(large_image)
```
