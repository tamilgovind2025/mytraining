# polygon draw

```python
import numpy as np
import cv2
import matplotlib.pyplot as plt

# Create an empty image with shape (512, 512, 3)
blank_img = np.zeros((512, 512, 3), dtype=np.uint8)

# Define points for the polygon
pts = np.array([[300, 150], [200, 300], [400, 300]], np.int32)

# Reshape the points array
pts = pts.reshape((-1, 1, 2))

# Draw polygon
cv2.polylines(blank_img, [pts], isClosed=True, color=(255, 0, 255), thickness=8)

# Display the image
plt.imshow(blank_img)
plt.show()
```

# adding text in images

```python
import numpy as np
import cv2

# Create a black image
image = np.zeros((512, 512, 3), np.uint8)

# Define text and font properties
text = "Hello, OpenCV!"
font = cv2.FONT_HERSHEY_SIMPLEX
font_scale = 1
font_color = (255, 255, 255)  # White color
thickness = 2
line_type = cv2.LINE_AA  # Anti-aliased line

# Get text size to determine text position
text_size = cv2.getTextSize(text, font, font_scale, thickness)[0]
text_width, text_height = text_size
text_x = (image.shape[1] - text_width) // 2  # Center text horizontally
text_y = (image.shape[0] + text_height) // 2  # Center text vertically

# Write text on the image
cv2.putText(image, text, (text_x, text_y), font, font_scale, font_color, thickness, line_type)

# Display the image
cv2.imshow('Text on Image', image)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

# chat gpt code

```python
import numpy as np
import cv2

# Create an empty image with shape (512, 512, 3)
image = np.zeros((512, 512, 3), dtype=np.uint8)

# Draw rectangle corners
cv2.rectangle(image, (50, 50), (100, 100), (255, 255, 255), 2)  # Top-left corner
cv2.rectangle(image, (412, 50), (462, 100), (255, 255, 255), 2)  # Top-right corner
cv2.rectangle(image, (50, 412), (100, 462), (255, 255, 255), 2)  # Bottom-left corner
cv2.rectangle(image, (412, 412), (462, 462), (255, 255, 255), 2)  # Bottom-right corner

# Draw rectangle center
cv2.rectangle(image, (231, 231), (281, 281), (255, 255, 255), 2)

# Draw empty circle
cv2.circle(image, (256, 256), 50, (255, 255, 255), 2)

# Draw filled circle
cv2.circle(image, (256, 256), 20, (255, 255, 255), -1)

# Draw diagonal line
cv2.line(image, (0, 0), (511, 511), (255, 255, 255), 2)

# Draw polygon
pts = np.array([[300, 150], [200, 300], [400, 300]], np.int32)
pts = pts.reshape((-1, 1, 2))
cv2.polylines(image, [pts], True, (255, 255, 255), 2)

# Add text
font = cv2.FONT_HERSHEY_SIMPLEX

cv2.putText(image, 'drawing images', (10, 30), font, 1, (255, 255, 255), 2, cv2.LINE_AA)
cv2.putText(image, 'govind', (10, 490), font, 0.5, (255, 255, 255), 1, cv2.LINE_AA)

# Display the image
cv2.imshow('Shapes', image)
cv2.waitKey(0)
cv2.destroyAllWindows()
```
