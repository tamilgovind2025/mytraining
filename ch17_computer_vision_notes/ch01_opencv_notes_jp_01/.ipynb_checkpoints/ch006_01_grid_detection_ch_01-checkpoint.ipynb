# grid detection 

```python
import cv2
import matplotlib.pyplot as plt 
%matplotlib inline 
import numpy as np 
```

```python
flat_chess=cv2.imread("DATA/flat_chessboard.png")
flat_chess=cv2.cvtColor(flat_chess,cv2.COLOR_BGR2RGB)
plt.imshow(flat_chess)
```

```python
found,corners=cv2.findChessboardCorners(flat_chess,(7,7))
```

```python
found
```

```python
#corners
```

```python
cv2.drawChessboardCorners(flat_chess,(7,7),corners,found) #inplace true
plt.imshow(flat_chess)
```

# dots images

```python
dots=cv2.imread("DATA/dot_grid.png")
dots=cv2.cvtColor(dots,cv2.COLOR_BGR2RGB)
plt.imshow(dots)
```

```python
found,corners=cv2.findCirclesGrid(dots,(10,10),cv2.CALIB_CB_SYMMETRIC_GRID)
```

```python
found
```

```python
#corners 
cv2.drawChessboardCorners(dots,(10,10),corners,found)
plt.imshow(dots)
```

# contour detection or edge detection 

```python
import cv2
import matplotlib.pyplot as plt 
import numpy as np 

%matplotlib inline 
```

```python
img=cv2.imread("DATA/internal_external.png",0)
img.shape
```

```python
plt.imshow(img,cmap="gray")
```

```python
contours,hier=cv2.findContours(img,cv2.RETR_CCOMP,cv2.CHAIN_APPROX_SIMPLE)
type(contours)
```

```python
len(contours)
```

```python
type(hier)
```

# external contour drawing

```python
external_contour=np.zeros(img.shape)
external_contour.shape
```

```python
for i in range(len(contours)):
    if hier[0][i][3]==-1:
        cv2.drawContours(external_contour,contours,i,255,-1)
plt.imshow(external_contour,cmap="gray")
```

# internal contour drawing 

```python
internal_contour=np.zeros(img.shape)
internal_contour.shape

for i in range(len(contours)):
    if hier[0][i][3]!=-1:
        cv2.drawContours(internal_contour,contours,i,255,-1)
plt.imshow(internal_contour,cmap="gray")
```

# particular contour drawing

```python
internal_contour=np.zeros(img.shape)
internal_contour.shape

for i in range(len(contours)):
    if hier[0][i][3]==4:
        cv2.drawContours(internal_contour,contours,i,255,-1)
plt.imshow(internal_contour,cmap="gray")
```

```python
internal_contour=np.zeros(img.shape)
internal_contour.shape

for i in range(len(contours)):
    if hier[0][i][3]==0:
        cv2.drawContours(internal_contour,contours,i,255,-1)
plt.imshow(internal_contour,cmap="gray")
```
