# ode

```python
import numpy as np
from scipy.integrate import solve_ivp
import matplotlib.pyplot as plt

# Define the ODE dy/dx = -2y
def ode(t, y):
    return -2 * y

# Initial condition y(0) = 1
y0 = [1]

# Time points for which to solve the ODE
t_span = (0, 5)

# Solve the ODE
sol = solve_ivp(ode, t_span, y0, t_eval=np.linspace(0, 5, 100))

# Plot the solution
plt.figure()
plt.plot(sol.t, sol.y[0], label='y(t)')
plt.xlabel('t')
plt.ylabel('y')
plt.legend()
plt.show()
```

# signal processing 


```python
import matplotlib.pyplot as plt

from scipy.signal import butter, filtfilt
import numpy as np

# Generate a noisy signal
t = np.linspace(0, 1, 100)
signal = np.sin(2*np.pi*5*t) + np.random.normal(size=t.size)

# Filter the signal
b, a = butter(4, 0.1)
filtered_signal = filtfilt(b, a, signal)

# Plot the original and filtered signals
plt.figure()
plt.plot(t, signal, label='Original Signal')
plt.plot(t, filtered_signal, label='Filtered Signal')
plt.legend()
plt.show()
```

# interpolation 

```python
import matplotlib.pyplot as plt
from scipy.interpolate import interp1d
import numpy as np

# Original data points
x = np.array([0, 1, 2, 3, 4])
y = np.array([0, 1, 4, 9, 16])

# Linear interpolation
f = interp1d(x, y, kind='linear')

# Interpolated points
x_interp = np.linspace(0, 4, 100)
y_interp = f(x_interp)

# Plot original data and interpolated function
plt.figure()
plt.plot(x, y, 'o', label='Original Data')
plt.plot(x_interp, y_interp, label='Interpolated Function')
plt.xlabel('x')
plt.ylabel('y')
plt.title('Linear Interpolation')
plt.legend()
plt.grid(True)
plt.show()
```

# optimization 

```python
import matplotlib.pyplot as plt
from scipy.optimize import minimize

def f(x):
    return (x - 2)**2

# Find the minimum of the function
result = minimize(f, 0)
min_x = result.x

# Generate x values for plotting
x = np.linspace(-1, 5, 100)
y = f(x)

# Plot the function and the minimum
plt.figure()
plt.plot(x, y, label='f(x) = (x - 2)^2')
plt.plot(min_x, f(min_x), 'ro', label='Minimum')
plt.xlabel('x')
plt.ylabel('f(x)')
plt.title('Minimization of f(x)')
plt.legend()
plt.grid(True)
plt.show()
```
