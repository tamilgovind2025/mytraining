# plotting the graphs

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Load the dataset
#url = "https://raw.githubusercontent.com/jbrownlee/Datasets/master/pima-indians-diabetes.data.csv"
#names = ['preg', 'plas', 'pres', 'skin', 'test', 'mass', 'pedi', 'age', 'class']
data = pd.read_csv("diabetes.csv")
data.head()
```

```python
# Pairplot for all features colored by 'Outcome'
sns.pairplot(data, hue='Outcome', diag_kind='hist')
plt.show()
```

```python
# Boxplot for each feature colored by "Outcome"
plt.figure(figsize=(12, 6))
sns.boxplot(data=data, orient='h', palette='Set2')
plt.show()
```

```python
# Correlation matrix heatmap
plt.figure(figsize=(10, 8))
sns.heatmap(data.corr(), annot=True, cmap='coolwarm')
plt.show()
```

```python
#bar plot 

plt.figure(figsize=(6, 4))
sns.countplot(data['Outcome'])
plt.show()
```

```python
#histogram 

plt.figure(figsize=(6, 4))
sns.histplot(data['Age'], bins=20, kde=True)
plt.show()
```

```python
#scatter plot 

plt.figure(figsize=(6, 4))
sns.scatterplot(x='Age', y='BloodPressure', hue='Outcome', data=data)
plt.show()
```

```python
#violin plot
plt.figure(figsize=(8, 6))
sns.violinplot(x='Outcome', y='Age', data=data, palette='Set2',hue="Outcome",legend=False)
plt.show()
```

```python
#box plot

plt.figure(figsize=(12, 6))
sns.boxplot(data=data, x='Outcome', y='Age')
plt.show()
```
