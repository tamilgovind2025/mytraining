# linear classification 

```python
#importing modules 

import numpy as np 
import pandas as pd 
import matplotlib.pyplot as plt 
```

```python
#load datasets 
dataset=pd.read_csv("Social_Network_Ads.csv")
x=dataset.iloc[:,:-1].values
y=dataset.iloc[:,-1:].values
#x
```

```python
#train and test split 
from sklearn.model_selection import train_test_split
xtrain,xtest,ytrain,ytest=train_test_split(x,y,test_size=0.25,random_state=0)
```

```python
print(x.shape)
print(xtrain.shape)
print(xtest.shape)
```

```python
#feature scalling 
from sklearn.preprocessing import StandardScaler
sc=StandardScaler()
xtrain_sc=sc.fit_transform(xtrain)
xtest_sc=sc.transform(xtest)
```

```python
#print(xtrain_sc)
#print(xtest_sc)
```

```python
from sklearn.linear_model import LogisticRegression
classifier=LogisticRegression()
classifier.fit(xtest_sc,ytest.ravel())
```

```python
#predict new result 
print(classifier.predict(sc.transform([[30,87000]])))
```

```python
#predict new result for single data 
y_pred=classifier.predict(xtest_sc)
y_pred
```

```python
#predict test datasets 
print(np.concatenate((y_pred.reshape(len(y_pred),1),(ytest.reshape(len(ytest),1))),1))
```

```python
#confusion matrix 
from sklearn.metrics import confusion_matrix, accuracy_score
cm=confusion_matrix(ytest,y_pred)
print(cm)
```

```python
#accuracy score 
accuracy_score(ytest,y_pred)*100
```

```python
#visuvalize training test 

from matplotlib.colors import ListedColormap
import numpy as np
import matplotlib.pyplot as plt

x_set, y_set = xtrain, ytrain
X1, X2 = np.meshgrid(np.arange(start=x_set[:, 0].min() - 10, stop=x_set[:, 0].max() + 10, step=0.25),
                     np.arange(start=x_set[:, 1].min() - 1000, stop=x_set[:, 1].max() + 1000, step=0.25))

plt.contourf(X1, X2, classifier.predict(sc.transform(np.array([X1.ravel(), X2.ravel()]).T)).reshape(X1.shape),
             alpha=0.75, cmap=ListedColormap(("red", "yellow")))

plt.xlim(X1.min(), X1.max())
plt.ylim(X2.min(), X2.max())

for i, j in enumerate(np.unique(y_set)):
    indices = np.where(y_set == j)[0]
    plt.scatter(x_set[indices, 0], x_set[indices, 1], c=ListedColormap(("gray", "blue"))(i), label=j)


plt.title("Logistic Regression")
plt.ylabel("Estimated Salary")
plt.xlabel("Age")
plt.legend()
plt.show()
```

```python
#visuvalize testing test 

from matplotlib.colors import ListedColormap
import numpy as np
import matplotlib.pyplot as plt

x_set, y_set = xtest, ytest
X1, X2 = np.meshgrid(np.arange(start=x_set[:, 0].min() - 10, stop=x_set[:, 0].max() + 10, step=1),
                     np.arange(start=x_set[:, 1].min() - 1000, stop=x_set[:, 1].max() + 1000, step=1))

plt.contourf(X1, X2, classifier.predict(sc.transform(np.array([X1.ravel(), X2.ravel()]).T)).reshape(X1.shape),
             alpha=0.75, cmap=ListedColormap(("red", "yellow")))

plt.xlim(X1.min(), X1.max())
plt.ylim(X2.min(), X2.max())

for i, j in enumerate(np.unique(y_set)):
    indices = np.where(y_set == j)[0]
    plt.scatter(x_set[indices, 0], x_set[indices, 1], c=ListedColormap(("gray", "blue"))(i), label=j)


plt.title("Logistic Regression")
plt.ylabel("Estimated Salary")
plt.xlabel("Age")
plt.legend()
plt.show()
```
