# svm regression 

```python
#importing  modules 

import numpy as np 
import matplotlib.pyplot as plt 
import pandas as pd 
```

```python
#reading datasets 
dataset=pd.read_csv("Position_Salaries.csv")
#dataset
x=dataset.iloc[:,1:-1].values
#x
y=dataset.iloc[:,-1:].values
#y
```

```python
#feature scalling

from sklearn.preprocessing import StandardScaler

sc_x=StandardScaler()
sc_y=StandardScaler()

x_sc_x=sc_x.fit_transform(x)
y_sc_y=sc_y.fit_transform(y)
```

```python
#no need here train and test 
#train the model 
from sklearn.svm import SVR

regressor=SVR(kernel='rbf')
regressor.fit(x_sc_x,y_sc_y.ravel())
```

```python
#predict the result 
y_predict =regressor.predict(sc_x.fit_transform([[6.5]]))
#y_predict  #now take inverse trasnform 
y_predict=y_predict.reshape(-1,1)
#y_predict
y_predict_acc=sc_y.inverse_transform(y_predict)
print(y_predict_acc.ravel()[0])
```

```python
#visualize svr result 

plt.scatter(x,y,color="red")
plt.plot(x,sc_y.inverse_transform( regressor.predict(x_sc_x).reshape(-1,1)),color="blue")
plt.title("SVM Regressions")
plt.xlabel("roles")
plt.ylabel("salary")
plt.show()
```

```python
#visualize svr result at high resolution 
x_grid=np.arange(np.min(x),np.max(x),0.1)
#print(x_grid)
x_grid=x_grid.reshape(len(x_grid),1)
#print(x_grid)
plt.scatter(x,y,color="red")
plt.plot(x_grid,sc_y.inverse_transform( regressor.predict(sc_x.fit_transform(x_grid)).reshape(-1,1)),color="blue")
plt.title("SVM Regressions")
plt.xlabel("roles")
plt.ylabel("salary")
plt.show()
```
