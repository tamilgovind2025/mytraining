# selection 

```python
DataFrame.at[]:
Accesses a single value by row and column labels.
It's faster than DataFrame.loc[] when accessing a single value.
Provides label-based indexing.
Syntax: DataFrame.at[row_label, column_label]
```

```python
import pandas as pd

data = {'A': [1, 2, 3], 'B': [4, 5, 6], 'C': [7, 8, 9]}
df = pd.DataFrame(data, index=['X', 'Y', 'Z'])
print(df)
value_at_Y_B = df.at['Y', 'B']
print(value_at_Y_B)  
```

```python
DataFrame.iat[]:
Accesses a single value by row and column index positions.
It's faster than DataFrame.iloc[] when accessing a single value.
Provides integer-based indexing.
Syntax: DataFrame.iat[row_index, column_index]
```

```python
value_iat_1_2 = df.iat[2,1]
print(value_iat_1_2) 
```
