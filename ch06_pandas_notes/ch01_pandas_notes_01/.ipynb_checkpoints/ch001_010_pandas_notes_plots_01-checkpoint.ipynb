# line plots

```python
import pandas as pd
import matplotlib.pyplot as plt

# Create a DataFrame
data = {'x': range(10), 'y': [i**2 for i in range(10)]}
df = pd.DataFrame(data)

# Plot a line chart
df.plot(x='x', y='y', kind='line')
plt.title('Line Plot')
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.show()
```

# histogram

# box plot

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

# Create a DataFrame
data = {'values': [np.random.normal(0, std, 100) for std in range(1, 4)]}
df = pd.DataFrame(data)

# Plot a box plot
plt.boxplot(df['values'])
plt.title('Box Plot')
plt.ylabel('Value')
plt.show()
```

# area plot

```python
import pandas as pd
import matplotlib.pyplot as plt

# Create a DataFrame
data = {'x': range(10), 'y1': [i**2 for i in range(10)], 'y2': [i**3 for i in range(10)]}
df = pd.DataFrame(data)

# Plot an area chart
df.plot(x='x', y=['y1', 'y2'], kind='area')
plt.title('Area Plot')
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.show()
```

# pie chart

```python
import pandas as pd
import matplotlib.pyplot as plt

# Create a DataFrame
data = {'category': ['A', 'B', 'C', 'D'], 'values': [20, 30, 40, 10]}
df = pd.DataFrame(data)

# Plot a pie chart
df.plot(y='values', kind='pie', labels=df['category'], autopct='%1.1f%%', legend=False)
plt.title('Pie Chart')
plt.ylabel('')
plt.show()
```

# 3d graph plot

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D

# Create a DataFrame with random data
data = pd.DataFrame(np.random.randn(100, 3), columns=['x', 'y', 'z'])

# Plot a 3D scatter plot
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')
ax.scatter(data['x'], data['y'], data['z'])
plt.title('3D Scatter Plot')
ax.set_xlabel('X-axis')
ax.set_ylabel('Y-axis')
ax.set_zlabel('Z-axis')
plt.show()
```

# stacked bar plot

```python
import pandas as pd
import matplotlib.pyplot as plt

# Create a DataFrame
data = {'category': ['A', 'B', 'C'],
        'value1': [10, 20, 30],
        'value2': [5, 15, 25]}
df = pd.DataFrame(data)

# Plot a stacked bar chart
df.set_index('category').plot(kind='bar', stacked=True)
plt.title('Stacked Bar Plot')
plt.xlabel('Category')
plt.ylabel('Values')
plt.show()
```

# Area Plot with Stacked Data:

```python
import pandas as pd
import matplotlib.pyplot as plt

# Create a DataFrame
data = {'x': range(10),
        'y1': [i**2 for i in range(10)],
        'y2': [i**3 for i in range(10)]}
df = pd.DataFrame(data)

# Plot a stacked area chart
df.plot(x='x', y=['y1', 'y2'], kind='area', stacked=True)
plt.title('Stacked Area Plot')
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.show()
```

# polar plot

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

# Create a DataFrame
data = {'theta': np.linspace(0, 2*np.pi, 100),
        'r': np.sin(np.linspace(0, 2*np.pi, 100))}
df = pd.DataFrame(data)

# Plot a polar plot
plt.figure(figsize=(6, 6))
ax = plt.subplot(111, projection='polar')
ax.plot(df['theta'], df['r'])
ax.set_title('Polar Plot')
plt.show()
```

# error bar plot 

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

# Create a DataFrame with random data
data = pd.DataFrame({
    'x': np.arange(10),
    'y': np.sin(np.arange(10)),
    'yerr': np.random.rand(10) * 0.5  # Random error bars
})

# Plot a line chart with error bars
plt.errorbar(data['x'], data['y'], yerr=data['yerr'], fmt='-o')
plt.title('Error Bars Plot')
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.show()
```

# donut chart

```python
import pandas as pd
import matplotlib.pyplot as plt

# Create a DataFrame
data = {'category': ['A', 'B', 'C', 'D'],
        'values': [25, 35, 20, 20]}
df = pd.DataFrame(data)

# Plot a donut chart
plt.pie(df['values'], labels=df['category'], autopct='%1.1f%%', startangle=90, wedgeprops=dict(width=0.3))
plt.title('Donut Chart')
plt.axis('equal')  # Equal aspect ratio ensures that pie is drawn as a circle
plt.show()
```

# stream plot 


```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

# Create a DataFrame
x = np.linspace(-3, 3, 100)
y = np.linspace(-3, 3, 100)
X, Y = np.meshgrid(x, y)
U = -1 - X**2 + Y
V = 1 + X - Y**2
df = pd.DataFrame({'X': X.flatten(), 'Y': Y.flatten(), 'U': U.flatten(), 'V': V.flatten()})

# Plot a streamplot
plt.streamplot(x, y, U, V)
plt.title('Streamplot')
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.show()
```

# waterfall chart 

```python
import pandas as pd
import matplotlib.pyplot as plt

# Create a DataFrame
data = {'category': ['Start', 'Step 1', 'Step 2', 'Step 3', 'End'],
        'values': [100, -20, -30, 50, 100]}
df = pd.DataFrame(data)

# Calculate cumulative sum for waterfall chart
df['cumulative_values'] = df['values'].cumsum()

# Plot a waterfall chart
plt.bar(df['category'], df['cumulative_values'], color=(df['values'] > 0).map({True: 'g', False: 'r'}))
plt.title('Waterfall Chart')
plt.xlabel('Category')
plt.ylabel('Values')
plt.show()
```
